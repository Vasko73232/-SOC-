```python
import pandas as pd
import numpy as np
import time
from sklearn.linear_model import SGDClassifier

from sklearn import model_selection
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
df_train = pd.read_csv('df_trainKDD.csv')
df_test = pd.read_csv('df_testKDD.csv')
```

```python
x_train = df_train.iloc[0:,1:42]
x_test = df_test.iloc[0:,1:42]
```

```python
y_train = df_train['class']
y_test = df_test['class'].values
```

```python
x_train.drop(labels = [107543],axis = 0, inplace = True)
```

```python
y_train.drop(labels = [107543],axis = 0, inplace = True)
x_train.drop(columns = ["id"],axis = 1)
```

```python
models = np.array([RandomForestClassifier(max_depth=5),KNeighborsClassifier(),SGDClassifier(loss="hinge", penalty="l2", max_iter=5),GaussianNB(),LogisticRegression()])
models_name = np.array(["RandForest","KNeighbors","SGDClas","GausNB","LogReg"])
i = 0;

df = pd.DataFrame(data=None, index=None, columns=["Name","Accuracy","Precision","recall","F1Score","Time"], dtype=None, copy=False)
df.head()
for model in models: 
  start_time = time.time()
  kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=24)
  accuracy = cross_val_score(model, x_train, y_train, cv=kf, scoring='accuracy').mean()
  precision = cross_val_score(model, x_train, y_train, cv=kf, scoring='precision').mean()
  recall = cross_val_score(model, x_train, y_train, cv=kf, scoring='recall').mean()
  f1 = cross_val_score(model, x_train, y_train, cv=kf, scoring='f1_weighted').mean()
  delta = time.time() - start_time
  df = df.append({'Name':models_name[i], 'Accuracy':accuracy, 'Precision':precision, 'recall':recall,"F1Score":f1,"Time":delta},ignore_index=True)
  i += 1
df.head()
```
